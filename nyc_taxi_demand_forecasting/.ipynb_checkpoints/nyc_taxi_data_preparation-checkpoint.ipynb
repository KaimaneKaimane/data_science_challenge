## Load Data

```python
file_path_list = [os.path.join(DATA_PATH, '2016_Yellow_Taxi_Trip_Data_1h.csv'),
                  os.path.join(DATA_PATH, '2017_Yellow_Taxi_Trip_Data_1h.csv')]

frames = list()
for file_path in file_path_list:
    frames.append(pd.read_csv(file_path))

data = pd.concat(frames)

outlier = pd.read_csv(os.path.join(DATA_PATH, 'outlier.csv'))
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'])
```

```python
data = data.sort_values(by='timestamp').reset_index(drop=True)
```

```python
for topic in weather_data_topics:
    print(topic)
    weater_data = pd.read_csv(f'/Users/clemenshensen/Projekte/data-science-interview/datasets/nyc taxi data/{topic}.csv')
    weater_data['datetime'] = pd.to_datetime(weater_data['datetime'])

    data[topic] = data['timestamp'].map(weater_data.set_index('datetime')['New York'])
```
